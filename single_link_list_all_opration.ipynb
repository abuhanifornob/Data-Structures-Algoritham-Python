```python
class Node:
  def __init__(self,data=None,next=None):
    self.data=data
    self.next=next

class Link_list:
  def __init__(self):
    self.head=None

  def print(self):  # ................print Operation
    if self.head is None:
      print("Link List is None")
      return
    iter=self.head
    llist=""
    while iter:
      llist+=str(iter.data) + "--> "
      iter=iter.next
    print(llist)
    
    # ,,,,,,,,,,,,,,,,,,, Length Operations ...................
  def get_length(self):
    if self.head is None:
      print("link List is Empty: ")
      return
    count=0
    iter=self.head
    while iter:
      count+=1
      iter=iter.next
    return count

 # .......................Insert at begging data.........................
    
  def insert_at_beggining(self,data):
    node=Node(data,self.head)
    self.head=node
    
    #.............................Insert at Endt point data......................
    
  def insert_at_end(self,data):
    if self.head is None:
      self.insert_at_beggining(data)
      return
    iter=self.head
    while iter.next:
      iter=iter.next
    node=Node(data,None)
    iter.next=node
    
    # ..........insert at user index .........................................
  def insert_at(self,index,data):
    if index>= self.get_length() or index<0:
      raise Exception("Invalide Index ")
      return
    if index==0:
      self.insert_at_beggining(data)
      return
    iter=self.head
    count=0
    while iter:
      if count==index-1:
        node=Node(data,iter.next)
        iter.next=node
        return
      iter=iter.next
        
  def remove_frist(self):
    self.head=self.head.next

  def remove_at(self,index):
    if index>= self.get_length() or index<0:
      raise Exception("Invalide Index ")
      return
    if index==0:
      self.remove_frist()
      return
    count=0
    iter=self.head
    while iter:
      if count==index-1:
        iter.next=iter.next.next
        break
      iter=iter.next
      count+=1
  def romove_end(self):
    iter=self.head
    count=0
    while iter:
      if count==self.get_length()-2:
        iter.next=iter.next.next
        break
      iter=iter.next
      count+=1
  def inser_value(self,data_list):
    self.head=None
    for data in data_list:
      self.insert_at_end(data)

  def insert_after_value(self, data_after, data_to_insert):
    if self.head is None:
      return
    count=0
    iter=self.head
    while iter:
      if iter.data==data_after:
        node=Node(data_to_insert,iter.next)
        iter.next=node
        break
      iter=iter.next
      count+=1
  def remove_by_value(self,data):
    if self.head is None:
      print("Link List Is Empty ")
      return
    if self.head.data==data:
      self.remove_frist()
      return
    iter=self.head
    while iter:
      if iter.next.data==data:
        iter.next=iter.next.next
        break
      iter=iter.next



        
if __name__=="__main__":
  ll=Link_list()
  ll.insert_at_beggining(50)
  ll.insert_at_beggining(40)
  ll.insert_at_beggining(40)
  ll.insert_at_beggining(40)
  ll.insert_at_beggining(500)
  ll.insert_at_beggining(600)
  ll.insert_after_value(600,1000)
  ll.remove_by_value(500)
  ll.print()
  print(ll.get_length())
```
