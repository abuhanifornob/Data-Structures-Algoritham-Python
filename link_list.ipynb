```python
class Node:
  def __init__(self,data=None,next=None):
    self.data=data
    self.next=next
class Link_list:
  def __init__(self):
    self.head=None
  def insert_at(self,index,data):
    if index<0 and index> self.get_length():
      raise Exception("Invalid Index")
    if index==0:
      self.insert_at_begining(data)
      return
    count=0
    iter=self.head
    while iter:
      if count==index-1:
        node=Node(data,iter.next)
        iter.next=node
        break
      iter=iter.next
      count+=1


  def get_length(self):
    if self.head is None:
      print('Link List is Empty')
      return
    count=0
    iter=self.head
    while iter is not None:
      iter=iter.next
      count+=1
    return count

  def insert_at_begining(self,data):
    if self.head is None:
      self.head=Node(data,None)
      return
    node=Node(data,self.head)
    self.head=node
  def insert_at_end(self,data):
    if self.head is None:
      self.head=Node(data,None)
      return
    iter=self.head
    while iter.next:
      iter=iter.next
    iter.next=Node(data,None)
  
  def remove_at(self,index):
    if index<0 and index> self.get_length():
      raise Exception ("Invalid Index")
      return
    if index==0:
      self.head=self.head.next
      return
    iter=self.head
    count=0
    while iter:
      if count==index-1:
        iter.next=iter.next.next
        break
      count+=1
      iter=iter.next 
    
  def print(self):
    if self.head is None:
      print("List is Empty")
    iter=self.head
    llink=''
    while iter is not None:
      llink +=str(iter.data)+" --> " if iter.next else str(iter.data)
      iter=iter.next
    print(llink)
    
  def insert_value(self,list):
    for data in list:
      self.insert_at_end(data)
  


if __name__=="__main__":
  ll=Link_list()
  ll.insert_at_begining(50)
  ll.insert_at_begining(50)
  ll.insert_at_begining(50)
  ll.insert_at_end(100)
  ll.insert_at(1,500)
  ll.remove_at(2)
  ll.insert_value(["banana","mango","grapes","orange"])
  print(ll.get_length())
  ll.print()
```
